```python
from shapely.geometry import Polygon
from matplotlib import pyplot as plt
import random
```

```python
p1,p2,p3,p4,p5=[(0,0),(1,0),(2,0),(1,1),(5,6)]
poly=Polygon([p1,p2,p3,p4,p5])
poly=poly.convex_hull
x,y=poly.exterior.xy
plt.plot(x,y)
plt.show()
```

```python
class polygon:
    def __init__(self,n):
        self.poly=Polygon()
        self.sides=n
        self.range=50
        if(n*10>self.range):
            self.range=n*10
    def generate(self):
        random.SystemRandom()
        x=random.sample(range(-self.range,self.range),self.sides)
        y=random.sample(range(-self.range,self.range),self.sides)
        z=list(zip(x,y))
        self.poly=Polygon(z)
        self.poly=self.poly.convex_hull
        n=len(self.poly.exterior.xy[0])-1
        while n<self.sides:
            random.SystemRandom()
            x,y=(random.randint(-self.range,self.range),random.randint(-self.range,self.range))
            x1,y1=self.poly.exterior.xy
            x1.append(x)
            y1.append(y)
            z=list(zip(x1,y1))
            self.poly=Polygon(z)
            self.poly=self.poly.convex_hull
            n=len(self.poly.exterior.xy[0])-1
    def plot(self):
        x,y=self.poly.exterior.xy
        plt.plot(x,y)
        plt.show()
```

```python
p=polygon(5)
p.generate()
print(p.poly.exterior.coords[:])
p.plot()
```
